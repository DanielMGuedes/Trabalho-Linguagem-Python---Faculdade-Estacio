```
# TRABALHO - PÓS GRADUAÇÃO EM CIÊNCIA DE DADOS E BIG DATA ANALYTICS
# Disciplina: Linguagem Python

import pickle
import os

arquivo = open('file.db', 'wb')
for i in range(10):
    pickle.dump(i, arquivo)
arquivo.close()
print ('Arquivo criado com sucesso')


db = {}

def show_menu():
    os.system('')
    print('Existe ' + str(len(db)) + ' Candidatos Registrados na Base\n')
    print('Escolha uma opção')
    print('1 - Cadastrar um novo Candidato')
    print('2 - Pesquisar um Candidato')
    print('3 - Modificar cadastro do Candidato')
    print('4 - Excluir Candidato')
    print('5 - Listar Todos os Candidatos')
    print('6 - Sair')
    return int(input(''))


def existe_registro(codigo):
    duplicated = False
    if codigo in db.keys():
        duplicated = True

    return duplicated


def include():
    candidato = {}

    codigo = input('Digite o nome do Candidato\n')

    if existe_registro(codigo):
        print('Já existe um Candidato cadastrado com este código, digite outro')
        include()
    else:
        candidato['codigo'] = codigo
        candidato['candidato'] = input('Digite o Código do candidato: \n')
        candidato['cargo'] = input('Informe o Cargo do candidato: \n')
        candidato['regiao'] = str(input('Informe a Região do candidato: \n'))
        candidato['votos'] = int(input('Informe a quantidade de votos do candidato: \n'))

        db[codigo] = candidato

        __main__()


def show_formatted_candidato(codigo):
    print('Código: ' + db[codigo]['codigo'])
    print('Nome: ' + db[codigo]['candidato'])
    print('Cargo: ' + db[codigo]['cargo'])
    print('Região: ' + str(db[codigo]['regiao']))
    print('Votos: ' + str(db[codigo]['votos']))


def show():
    codigo = input('Digite o nome do candidato:\n')
    if existe_registro(codigo):
        show_formatted_candidato(codigo)
        trash = input('pressione ENTER para voltar ao menu\n')
        __main__()
    else:
        print('Registro não encontrado')
        trash = input('pressione ENTER para voltar ao menu\n')
        __main__()


def modify():
    candidato = {}

    codigo = input('Digite o nome do candidato:\n')

    if existe_registro(codigo):
        candidato['codigo'] = codigo
        candidato['candidato'] = input('Digite o nome do candidato: \n')
        candidato['cargo'] = input('Informe o Cargo do candidato: \n')
        candidato['regiao'] = int(input('Informe a Região do candidato: \n'))
        candidato['votos'] = float(input('Informe a quantidade de votos do candidato: \n'))

        db[codigo] = candidato

        __main__()
    else:
        print('Registro não encontrado')
        trash = input('pressione ENTER para voltar ao menu\n')
        __main__()


def delete():
    codigo = input('Digite o nome do candidato:\n')

    if existe_registro(codigo):
        del (db[codigo])
        print('Registro apagado com sucesso')
        trash = input('pressione ENTER para voltar ao menu\n')
        __main__()
    else:
        print('Registro não encontrado')
        trash = input('pressione ENTER para voltar ao menu\n')
        __main__()


def list():
    if len(db) > 0:
        for codigo in db.keys():
            show_formatted_candidato(codigo)
            print('-' * 10)
        trash = input('pressione ENTER para voltar ao menu\n')
        __main__()
    else:
        print('Não há registros para serem exibidos')
        trash = input('pressione ENTER para voltar ao menu\n')
        __main__()


def __main__():
    option = show_menu()
    if (option == 1):
        include()
    elif (option == 2):
        show()
    elif (option == 3):
        modify()
    elif (option == 4):
        delete()
    elif (option == 5):
        list()
    elif (option == 6):
        print('Saindo do programa')
        exit()
    else:
        print('Instrução não reconhecida, tente outra opção')
        __main__()


__main__()
```
